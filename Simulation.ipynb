```python
def pick_guess(remaining, constraints, unknown, guesses):
    """
    Pick a new guess from the remaining words given the current constraints
    and vocabulary setting.
    Args:

    remaining: a list of words that haven't been guessed yet.
    constraints: a dictionary of constraints on the target word, where the keys
                    are the positions of the target word and the values are sets
                    of possible letters for each position.
    v: the vocabulary setting, which is either 'common' or 'uncommon'.

    Returns:
    guess: the new guess.
    """
    # first choice
    if len(guesses)==0:
        choice = remaining.sample(n=1, weights=remaining['percentage'])
        guess = np.random.choice(['salet','reast','trace','crate','slate', choice.iloc[0,0]])
        return [guess, remaining, unknown]
    
    # filter based on the constraints
    correct = constraints[1]
    wrong_loc = constraints[0.5]
    incorrect = constraints[0]
    cor_loc = constraints['loc']

    filter_cor = remaining.copy(deep=True)


    for (value, index) in correct:
        filter_cor = filter_cor[filter_cor['word'].str[index] == value]
    
    #print(filter_cor)
        
    filter_new = filter_cor.copy(deep=True)
    
    for n in incorrect:
        filter_new = filter_new[~filter_new['word'].str.contains(n)]
    #print(constraints)
    for (value, index) in wrong_loc:
        pos_left = [0,1,2,3,4]
        for i in cor_loc:
            pos_left.remove(i)
        if index in pos_left:
            pos_left.remove(index)
        #print(pos_left, value, constraints)
        filter_new = filter_new[filter_new.apply(lambda x: any([x['word'][i] == value for i in pos_left]), axis=1)]
    if len(filter_new) > 0:
        #print(filter_new)
        filter_new['percentage'] = filter_new['count']/filter_new['count'].sum()
        choice = filter_new.sample(n=1, weights=filter_new['percentage'])
        return [choice.iloc[0,0], filter_new, unknown]
    
    else: 
        #print('i dont know')
        filter_cor1 = unknown.copy(deep=True)

        for (value, index) in correct:
            filter_cor1 = filter_cor1[filter_cor1['word'].str[index] == value]

        filter_new1 = filter_cor1.copy(deep=True)

        for n in incorrect:
            filter_new1 = filter_new1[~filter_new1['word'].str.contains(n)]
        
        for (value, index) in wrong_loc:
            pos_left = [0,1,2,3,4]
            for i in cor_loc:
                pos_left.remove(i)
            if index in pos_left:
                pos_left.remove(index)
            #print(pos_left, value, constraints)
            filter_new1 = filter_new1[filter_new1.apply(lambda x: any([x['word'][i] == value for i in pos_left]), axis=1)]
        filter_new1['percentage'] = filter_new1['count']/filter_new1['count'].sum()
        choice1 = filter_new1.sample(n=1)
        return [choice1.iloc[0,0], filter_new, unknown]
        
```

```python
def update_constraints(constraints, guess, target):
    """
    Update the constraints based on the new guess and target.

    Args:

    constraints: a dictionary of constraints on the target word, where the keys
                 are the positions of the target word and the values are sets
                 of possible letters for each position.
    guess: the new guess.
    target: the target word.

    Returns:
    constraints: the updated dictionary of constraints.
    """
    update_constraints = {
        1: [],
        0.5: [],
        'loc': []
    }

    update_constraints[0] = constraints[0]
    chars_target = list(target)
    chars_guess = list(guess)

    # If the guess is in the target word, update the constraints for all matching positions.
    for i, letter in enumerate(chars_guess):
        
        if letter not in target:
            if letter not in update_constraints[0]:            
                update_constraints[0].append(letter)
        else:
            if letter == chars_target[i]:
                chars_guess[i] = "."
                chars_target[i] = "-"
                update_constraints[1].append((letter, i))
                update_constraints['loc'].append(i)


    for i, letter in enumerate(chars_guess):
        if letter in chars_target:
            chars_guess[i] = "."
            chars_target[chars_target.index(letter)] = "-"
            update_constraints[0.5].append((letter, i))

    return update_constraints
```

```python
def play_wordle(target, words, v):
    """
    Play Wordle for a given target word.

    Args:

    target: the target word.
    words: the words available as guesses.
    v: the vocabulary setting. ranging from 0 to 1

    Returns:
    guesses: a list of guesses used to identify target otherwise None.
    """

    # Make a copy of the list of words to avoid modifying the original.
    remaining = list(words)

    # Initialize the guesses and constraints.
    guesses = []
    #0 is incorrect, 0.5 is correct but incorrect position, 1 is correct (a tuple consist of letter and index)
    constraints = {
        0:[],
        1: [],
        0.5: [],
        'loc': []
    }
    # assumption that v is the percentage that the simulation know v*100% of the most common words
    remaining = sorted_words.sample(frac=v, weights=sorted_words['percentage'])
    unknown = sorted_words[~sorted_words.word.isin(remaining.word)]
    # While there are still remaining words to guess...
    while True:

        # Pick a new guess from the remaining words given the current constraints
        # and vocabulary setting.
        guess_data = pick_guess(remaining, constraints, unknown, guesses)

        guess = guess_data[0]
        remaining = guess_data[1]
        unknown = guess_data[2]
        #print("this is my guess: " + guess)

        # Remove the new guess from the remaining words.
        remaining = remaining.drop(remaining[remaining['word'] == guess].index)

        # Add the new guess to the list of guesses.
        guesses.append(guess)
        #print(constraints, guess)
        # Update the constraints based on the new guess and target.
        constraints = update_constraints(constraints, guess, target)
        # If the new guess matches the target, we win!
        if guess == target:
            return guesses

    # If we get here, we've used up all the guesses without finding the target.
    return []
```
